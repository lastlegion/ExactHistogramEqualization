```python
####Power law transformations
Op_pw = I.copy()

gammas = [0.67, 0.57,0.47,0.37, 0.2, 0.1, 0.04]
for gamma in gammas:
    for i in range(0, I.width):
        for j in range(0, I.height):
            val = 255*((I[i,j][0]/255)**gamma)
            Op_pw[i,j] = (val,val,val)
    Op_pw.show()
```

```python
#Piecewise transformations
Oppw = I.copy()
sfactor = 255/(math.log(1+255))
threshs = [5,10,15,20,30,40,50]

for thresh in threshs:
    opthresh = sfactor*log(1+thresh)
    for i in range(0, I.width):
        for j in range(0, I.height):
            inval = I[i,j][0]
            if(inval <=thresh):
                opval = sfactor*log(1+inval) #Log
            else:
                opval = inval + opthresh #Linear
            Oppw[i,j] = (opval, opval, opval)
    Oppw.show()
```
